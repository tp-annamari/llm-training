## Control flow

### Number guessing game

The following exercise introduces basic control flow through a simple game. The program will think of a number between 1 and 100 and repeatedly prompt you to guess it, informing you if your guess is too low, too high, or correct!

````python
import random # import a module
# or
# from random import randint

secret = random.randint(1, 100) # pick a secret number
attempts = 0

print("I've thought of a number between 1 and 100.")
print("Try to guess it!")

# infinite loop
while True: # boolean values: True, False
    attempts += 1
    user_input = input(f"Attempt {attempts}: Enter your guess: ")

    # empty string is falsy, so it's checked with the `not` logical operator
    if not user_input:
        print("Empty input, try again!")
        attempts -= 1
        continue

    try:
        # variables are either global or function level
        guess = int(user_input) # convert input to integer
    except ValueError:
        print("Invalid input. Please enter a whole number.")
        attempts -= 1
    except EOFError:
        print(f"Quitting the game. The number was {secret}")
        break
    finally:
        pass # empty block is a syntax error, so we do nothing with pass

    if guess < 1 or guess > 100: # logical operators: or, and
        print("Please guess a number between 1 and 100.")
        attempts -= 1
    elif guess < secret: # elif instead of 'else if'
        print(f"Too low! ({guess})")
    elif guess > secret:
        print(f"Too high! ({guess})")
    else:
        print(f"Congratulations! You guessed the number {secret} correctly in {attempts} attempts! 🎉")
        break
````

## Data structures

### List - Hangman

````python
import random

print("Let's play Hangman! Enter custom words separated by spaces or leave it blank to use defaults")

# create a list
default_words = ["python", "java", "kotlin", "javascript", "ruby" "lisp"]

# strip() removes leading and trailing whitespaces
user_input = input("Enter custom words separated by spaces (optional): ").strip()

# conditional expression, also called as ternary operator
words = user_input.split() if user_input else default_words
print(words)

# select a random item
secret = random.choice(words)

# list comprehension
secret_letters = [c for c in secret if c.isalnum()]

# populate a list with the same item
blanks = ['_'] * len(secret_letters)

guessed = []
attempts = 6

# 'in' operator checks if a value is present in a collection
while "_" in blanks and attempts > 0:
    # multiline string
    print(f"""
Word:           {" ".join(blanks)}
Guessed so far: {guessed}
Attempts left:  {attempts}""", flush=True) # forcibly flush the stream
    
    guess = input("Enter a letter: ").strip()
    
    if len(guess) != 1 or not guess.isalnum() or guess in guessed:
        print("Invalid guess, try again.")
        continue
    
    guessed.append(guess)
    
    # 'not in' operator checks if a value is absent in a collection
    if guess not in secret_letters:
        attempts -= 1
        print("❌ Wrong!")
        continue

    # enumerate() returns a lazy sequence of (index, value) pairs
    for i, ch in enumerate(secret_letters):
        if ch == guess:
            blanks[i] = guess

if "_" not in blanks:
    print("\n🎉 You won! The word was", secret)
else:
    print("\n💥 Game over. The word was", secret)
````

### Dictionary - Capitals Quiz

````python
import random

# create a dict
capitals = {
    "France": "Paris",
    "Germany": "Berlin",
    "Japan": "Tokyo",
    "India": "New Delhi",
    "Canada": "Ottawa",
    "Australia": "Canberra",
    "Egypt": "Cairo"
}

score = 0
guesses = {} # empty brackets create dict, not set!
countries = list(capitals.keys()) # create list from dict keys
questions = min(5, len(countries)) # number of questions
random.shuffle(countries) # shuffle countries

for country in countries[:questions]: # slice of the first 'questions' items
    answer = input(f"What is the capital of {country}? ").strip()
    guesses[country] = answer # add item to dict
    if answer.lower() == capitals[country].lower(): # get value from dict
        print("✓ Correct!")
        score += 1
    else:
        print(f"✗ Wrong! Answer: {capitals[country]}")

print(f"Your final score: {score}/{questions}\n")

print("Summary of your answers:")
# iterate through keys and values
for country, answer in guesses.items():
    correct = capitals[country]
    mark = "✓" if answer.lower() == correct.lower() else "✗"
    print(f"  {country:>10}: {answer = }, {correct = } {mark}") # prettify the output

# dict comprehension
wrong = {c: g for c, g in guesses.items() if g.lower() != capitals[c].lower()}
correct = {c: g for c, g in guesses.items() if g.lower() == capitals[c].lower()}
print("\nWrong guesses:", wrong)
print("Correct guesses:", correct)
````
