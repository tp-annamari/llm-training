# Data Structures

Python provides a rich set of built-in data structures that allow you to efficiently store, organize, and manipulate data. Understanding these structures, such as lists, tuples, sets, and dictionaries, is fundamental to writing effective Python code. Each data structure has unique characteristics and use cases, making it important to choose the right one for your specific needs.

## List

Lists are mutable sequences, typically used to store collections of *homogeneous* items, defined with `[...]`:

````python
numbers = [1, 2, 3, 4, 5, 6]
numbers
````

Use the `del` statement to delete one or more items from a list:

````python
del numbers[0]
print(numbers)
````

````python
del numbers[2:4]
print(numbers)
````

*List comprehensions* provide a concise way to create lists that resemble *inverted for loops*. They are considered more *Pythonic* than using higher-order functions, such as `map()` and `filter()`, which are common in other programming langues.

````python
[i ** 2 for i in range(10)]
````

A list comprehension consists of one or more `for` clause and optional `if` clauses:

````python
[(x, y) for x in [1,2,3] for y in [3,1,4] if x != y]
````

It's equivalent to:

````python
combs = []
for x in [1,2,3]:
    for y in [3,1,4]:
        if x != y:
            combs.append((x, y))
combs
````

Learn about list methods: https://docs.python.org/3/tutorial/datastructures.html#more-on-lists

## Iterator

Although **higher-order functions are generally discouraged** in favor of comprehensions or simple loops, they are still present in the language. Similarly to `range()`, these functions (`map()`, `filter()`) are *lazy* and they return an *iterator*. Use the `lambda` keyword to create a *single-line* anonymous function:

````python
even_nums = filter(lambda x: x % 2 == 0, range(1, 11))
digits = map(str, even_nums)
digits
````

We need to *consume* an iterator to use its values:

````python
", ".join(digits)
````

Once an iterator is *exhausted*, it no longer provides any value:

````python
list(digits)
````

For more iterator functions, see the `itertools` module: https://docs.python.org/3/library/itertools.html

## Tuple

Tuples are *immutable* sequences, typically used to store collections of heterogeneous data. They are defined with `(...)` or just by using commas:

````python
t1 = (1, 2, 3)
t2 = 'a', 'b', 'c' # parentheses are optional
t3 = 4, # single-element tuple needs a trailing comma
print(t1, t2, t3)

try:
    t1[0] = 100
except TypeError as e:
    print("Cannot mutate:", e)
````

Assigning multiple variables on a single line creates a temporary tuple under the hood:

````python
point = (10, 20)
x, y = point # unpack
print(x, y)
````

````python
a, b = 5, 7
print(a, b)
a, b = b, a # swap two variables
print(a, b)
````

````python
first, *rest = "a", "b", "c", "d" # take first, pack the rest
print(first, rest)
````

We can return multiple values from a function using tuples under the hood:

````python
def min_max(seq):
    """Return (min, max) as a tuple."""
    return min(seq), max(seq)

low, high = min_max([8, 3, 10, 1])
print(low, high)
````

We can use comprehensions within parenthesis, but **it won't create a tuple**, instead it will return a **generator**:

````python
it = (i ** 2 for i in range(2 ** 1000)) # generator expression
print(it)
print('first item: ', next(it))
print('second item:', next(it))
````

## Set

Sets are unordered collections with no duplicate elements and `O(1)` lookup, defined with `{}`:

````python
{"apple", "orange", "apple", "pear", "orange", "banana"}
````

Use `set()` to create an empty set because `{}` creates a `dict`:

````python
print("{}:", type({}))
print("set():", set())
````

Similarly to list comprehensions, *set comprehensions* are also supported:

````python
duplicated_digits = ["1", "1", "1", "2", "2", "3", "11"]
{int(d) for d in duplicated_digits if len(d) == 1}
````

## Dictionary

Dictionaries are indexed by keys, which can be any *immutable type*, and map hashable values to arbitrary objects. They are defined with `{...}`, similar to sets, but they contain key-value pairs separated by a colon `:`. They support *dict comprehensions*:

````python
capitals = {
    "France": "Paris",
    "Germany": "Berlin",
    "Canada": "Ottawa",
    "Australia": "Canberra",
}
value_to_key = {v: k for k, v in capitals.items()}
print(value_to_key)
````

The `keys()`, `values()`, and `items()` methods return a *dynamic view* of the dictionary's contents. This means that if the dictionary changes, the view reflects these changes:

````python
keys = capitals.keys()
print(list(keys))
del capitals["France"]
print(list(keys))
````

We can use tuples as dictionary keys because they are immutable, unlike lists:

````python
coordinate = (52.1, 4.3)
{coordinate: "Amsterdam"}
````

## Slicing

While indexing is used to obtain individual value, slicing allows you to obtain a sub-collection *of the same type*:

````python
cities = ["Paris", "Berlin", "Tokyo", "Ottawa", "Canberra"]
cities = cities[1:-1] # omit the first and last items
print(cities)
````

The slice syntax resembles indexing and the parameters of `range()`. It takes 3 optional "arguments", which have the following default values if omitted:

- `start` = 0
- `stop` = number of items
- `step` = 1

Using negative values is the same as subtracting from the length:

````python
(s := 'abcdefghijkl') # walrus operator
````

````python
s[0:len(s):1] # default values
````

````python
s[2:] # skip first two
````

````python
s[:5] # take first five
````

````python
s[:len(s) - 1] # skip last
````

````python
s[:-1] # same as above
````

````python
s[-3:] # last three
````

````python
s[::2] # take every second item
````

````python
s[::-1] # reverse content
````

````python
s[:] # create a copy (useful for lists)
````
