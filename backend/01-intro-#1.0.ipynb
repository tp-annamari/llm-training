## Basic types

### String

Use *f-strings* to embed values in a string:

````python
name = "John"
print(f"Welcome, {name}!")
````

For more advanced formatting, see: https://docs.python.org/3/tutorial/inputoutput.html#formatted-string-literals

### Function

The built-in `print()` function accepts variadic arguments. Let's wrap it in our own function to see how it works:

````python
def greet(*names, sep=", ", end="!\n"):
    """Greet people by name, separated by sep and ending with end."""
    print("Hello", *names, sep=sep, end=end)

greet("Julia", "John", "Jeff")
````

The `*names` parameter *packs* the arguments into a single variable and we can use the same syntax to *unpack* it when passing it to the `print()` function.

The `"""triple quoted"""` string literal that occurs as the first statement of the function is a [docstring](https://peps.python.org/pep-0257/), and can be read at runtime:

````python
greet.__doc__
````

### Numbers

There are three distinct numeric types: `int`, `float`, and `complex`.

The standard library includes the additional numeric types `Fraction`, for rationals, and `Decimal`, for floating-point numbers with user-definable precision.

`int` has arbitrary precision, and we can produce very large numbers:

````python
2 ** 1000 # 2 to the power of 1000
````

Dividing with 0 always produces an error, even with `float` numbers:

````python
1.0 / 0.0
````

Infinity and NaN can be produced in the following way:

````python
float("inf"), float("-inf"), float("nan")
````
